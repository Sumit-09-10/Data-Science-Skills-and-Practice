# NumPy ML Challenge 33: Vectorization (No Loops)

## Dataset
ages = np.array([15, 18, 22, 35, 60, 12, 45, 28])

### Task 1
Increase every age by **1 year**.

### Task 2
Find all adults (**age >= 18**).

### Task 3
Create a binary label array.
Rules:

- Age < 18 → 0
- Age >= 18 → 1

### Task 4
Calculate the average age of adults only.

### Task 5
Calculate:

- Number of adults
- Number of minors

### Task 6 (ML-Oriented)
Create age groups using **np.where()**.

Rules:

- Age < 18 → "Minor"
- 18 <= Age <= 40 → "Adult"
- Age > 40 → "Senior"



````python
ages = np.array([15, 18, 22, 35, 60, 12, 45, 28])

new_ages=ages+1

mask_adults=ages>=18
adults=ages[mask_adults]

labels=np.where(ages<18, 0, 1)

avg_adult_age=np.mean(adults)

adult_count=len(adults)


minor_count=len(ages)-len(adults)

# Define your list of conditions in order
conditions = [(ages < 18),(ages >= 18) & (ages <= 40),(ages > 40)]
choices = ["Minor", "Adults", "Senior"]
age_group = np.select(conditions, choices, default="Unknown")




print("New Ages:")
print(new_ages)

print("\nAdults:")
print(adults)

print("\nLabels:")
print(labels)

print("\nAverage Adult Age:")
print(avg_adult_age)

print("\nAdult Count:")
print(adult_count)

print("\nMinor Count:")
print(minor_count)

print("\nAge Groups:")
print(age_group)
````

# NumPy ML Challenge 35: Broadcasting with Column Vectors

## Dataset

```python
import numpy as np

sales = np.array([
    [120, 150, 180],
    [200, 210, 220],
    [160, 170, 180],
    [300, 310, 320]
])

discount = np.array([
    [10],
    [20],
    [15],
    [30]
])
```

## Tasks

### Task 1
Subtract the discount from each row using broadcasting.

**Store in:** `final_sales`

### Task 2
Calculate the total sales of each store after the discount.

**Store in:** `store_total`

### Task 3
Calculate the average sales of each product after the discount.

**Store in:** `product_average`

### Task 4
Find the index of the store with the highest total sales.

**Store in:** `best_store`

### Task 5
Normalize each product (column-wise).

Formula:

`(final_sales - mean) / std`

**Store in:** `normalized_sales`

````python
sales = np.array([
    [120, 150, 180],
    [200, 210, 220],
    [160, 170, 180],
    [300, 310, 320]
])

discount = np.array([
    [10],
    [20],
    [15],
    [30]
])

final_sales=sales-discount
store_total=np.sum(final_sales, axis=1)
product_average=np.mean(final_sales, axis=0)
best_store=np.argmax(store_total)
avg=np.mean(final_sales, axis=0)
std_d=np.std(final_sales, axis=0)
normalized_sales=(final_sales-avg)/std_d

print(final_sales)
print(store_total)
print(product_average)
print(best_store)
print(normalized_sales)
````
