## Q15. Matrix Multiplication — dot vs @

**Question:** Given `A = [[1,2],[3,4]]` and `B = [[5,6],[7,8]]`, compute the matrix product using both `np.dot()` and the `@` operator, then verify they are equal.

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

dot_product = np.dot(A, B)
matmul      = A @ B

print("np.dot(A, B):\n", dot_product)
print("A @ B       :\n", matmul)
print("Both equal  :\n", dot_product == matmul)
```

## Q16. Determinant & Inverse

**Question:** Given `A = [[4,7],[2,6]]`, compute and print:
- The **determinant** using `np.linalg.det`
- The **inverse** using `np.linalg.inv`
- Verify by multiplying `A @ A_inv` (should give the identity matrix)

```python
A = np.array([[4, 7],
              [2, 6]])
det = np.linalg.det(A)
inv = np.linalg.inv(A)
identity_check = A @ inv

print("Determinant      :", round(det, 4))
print("Inverse:\n", inv)
print("A @ A_inv (≈ I):\n", np.round(identity_check, 4))
```

## Q18. Column-wise Normalisation

**Question:** Given a dataset of [height, weight] rows, compute:
- **Column mean** and **column std**
- **Normalised data** (Z-score normalisation column-wise)

```python
data = np.array([[170, 65],
                 [180, 80],
                 [175, 75],
                 [160, 55]])

avg_col = np.mean(data, axis=0)
std_col = np.std(data, axis=0)
norm    = (data - avg_col) / std_col

print("Column mean        :", avg_col)
print("Column std         :", std_col)
print("Normalised data:\n", np.round(norm, 4))
```

## Q19. Train / Test Split

**Question:** Create `data = np.arange(1, 21)` (20 elements). Split it so the **first 16** elements become training data and the **last 4** become test data, using both slicing and `np.split`.

```python
data = np.arange(1, 21)
# Method 1 — slicing
train_slice = data[:16]
test_slice  = data[16:]
print("Train (slicing):", train_slice)
print("Test  (slicing):", test_slice)

# Method 2 — np.split
train, test = np.split(data, [16])
print("Train (split)  :", train)
print("Test  (split)  :", test)
```

## Q20. Dataset Analysis — Height & Weight

**Question:** Given a 5-row dataset of `[height, weight]`:
1. Compute **mean per column**
2. Find the **max per column**
3. Filter rows where **weight > 70**
4. **Sort rows** by height (ascending)
5. **Z-score normalise** column-wise

```python
data = np.array([[170, 65],
                 [180, 80],
                 [175, 75],
                 [160, 55],
                 [185, 90]])

print("Mean per column          :", np.mean(data, axis=0))
print("Max  per column          :", np.max(data, axis=0))
print("Rows where weight > 70:\n", data[data[:, 1] > 70])

idx         = np.argsort(data[:, 0])
sorted_data = data[idx]
print("Sorted by height:\n", sorted_data)

norm = (data - np.mean(data, axis=0)) / np.std(data, axis=0)
print("Z-score normalised:\n", np.round(norm, 4))
```

## Q22. Handling NaN Values

**Question:** Given `marks = [85, NaN, 78, 92, NaN, 88]`:
1. Print the array with NaNs
2. **Detect** which positions are NaN using `np.isnan`
3. Compute the **mean ignoring NaNs** using `np.nanmean`
4. **Replace NaNs** with the mean using `np.nan_to_num`

```python
marks = np.array([85, np.nan, 78, 92, np.nan, 88])
print("Original array  :", marks)
print("Is NaN mask     :", np.isnan(marks))

mean_no_nan = np.nanmean(marks)
print("Mean (ignore NaN):", mean_no_nan)

filled = np.nan_to_num(marks, nan=mean_no_nan)
print("NaNs replaced   :", filled)
```

## Q29. Trace — Sum of Diagonal

**Question:** Given matrix `A = [[4,2,1],[0,5,3],[7,8,6]]`:
- Extract the **diagonal**
- Compute the **trace** using `np.trace`
- Verify that `trace == sum of diagonal`

```python
A = np.array([[4, 2, 1],
              [0, 5, 3],
              [7, 8, 6]])
diag  = np.diag(A)
trace = np.trace(A)
print("Diagonal     :", diag)
print("np.trace(A)  :", trace)
print("diag.sum()   :", diag.sum())
print("Equal?       :", trace == diag.sum())
```

## Q30. Matrix Rank

**Question:** Compute the **rank** of two matrices:
- `A = [[1,2],[2,4]]` (linearly dependent rows — rank 1)
- `B = [[1,2],[3,4]]` (linearly independent — rank 2)
- Print whether they have equal rank

```python
A = np.array([[1, 2], [2, 4]])
B = np.array([[1, 2], [3, 4]])

rank_A = np.linalg.matrix_rank(A)
rank_B = np.linalg.matrix_rank(B)

print("Rank of A:", rank_A)
print("Rank of B:", rank_B)
print("Equal rank?", rank_A == rank_B)
```

## Q31. Submatrix Extraction — np.ix_

**Question:** Given a 4×4 matrix, extract a **2×2 submatrix** formed by rows [0, 2] and columns [1, 3] using `np.ix_`.

```python
A = np.array([[11, 12, 13, 14],
              [21, 22, 23, 24],
              [31, 32, 33, 34],
              [41, 42, 43, 44]])
print("Full matrix:\n", A)

rows = [0, 2]
cols = [1, 3]
sub_matrix = A[np.ix_(rows, cols)]
print("Submatrix (rows 0,2 | cols 1,3):\n", sub_matrix)
```

## Q32. Row-wise Sorting — argsort + take_along_axis

**Question:** Given `A = [[30,10,20],[60,40,50]]`, sort each **row independently** in ascending order using `np.argsort` and `np.take_along_axis`.

```python
A = np.array([[30, 10, 20],
              [60, 40, 50]])
print("Original:\n", A)

idx           = np.argsort(A, axis=1)         # sort indices per row
sorted_array  = np.take_along_axis(A, idx, axis=1)  # reorder using those indices
print("Argsort indices:\n", idx)
print("Row-wise sorted:\n", sorted_array)
```
